```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split


dataset = pd.read_csv("./data/mall_customers.csv")
dataset.head()
```

Analizando las correlaciones de manera gr√°fica

```python
import matplotlib.pyplot  as plt
import seaborn as sns
sns.pairplot(dataset, height=2.5)
plt.show()
```

```python

x = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss= []
for i in range(1,11):
    kmeans= KMeans(n_clusters=i,init="k-means++",max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)   
plt.title("Metodo del codo")
plt.xlabel("N√∫mero de clusters")
plt.ylabel("wcss(k)")
plt.show()
```

```python
kmeans= KMeans(n_clusters=5,init="k-means++",max_iter=300, n_init=10, random_state=0)
kmeans.fit(x)

```

```python
y_kmeans = kmeans.fit_predict(x)
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans==0,1], s=100,c="red",label="Cluster 001")
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans==1,1], s=100,c="blue",label="Cluster 002")
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans==2,1], s=100,c="green",label="Cluster 003")
plt.scatter(x[y_kmeans == 3,0], x[y_kmeans==3,1], s=100,c="cyan",label="Cluster 004")
plt.scatter(x[y_kmeans == 4,0], x[y_kmeans==4,1], s=100,c="magenta",label="Cluster 005")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=300,c="yellow", label="Baricentros")
plt.title("Cluster de clientes")
plt.xlabel("Ingreso anuales en miles")
plt.ylabel("Puntuacion de gastos")
plt.legend()
plt.show()
```
