1. Preprocesado de datos

```python
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv("./data/salary-data.csv")

dataset.head()
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
print(dataset)
dataset.Salary.hist(bins=30)
```

Muestra la relacion entre el n√∫mero de registros y el los valores que hay entre ellos

```python
dataset.Salary.hist(bins=30)

```

```python
dataset.YearsExperience.hist(bins=30)
```

```python
import matplotlib.pyplot  as plt
import seaborn as sns
sns.pairplot(dataset, height=2.5)
plt.show()
```

2. Division de datos en entrenamiento y testing 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1 / 3, random_state=0)
```

3. Crear modelo de regresion lineal simple con el conjunto de entrenamiento 

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(x_train, y_train)
```

4. Predecir el conjunto de test

```python
y_pred = regression.predict(x_test)
```

5. Visualizar los datos de entrenamiento

```python
plt.scatter(x_train, y_train, color="red")
plt.plot(x_train, regression.predict(x_train), color="blue")
plt.title("Sueldo vs A√±os de experiencia (Conjunto de datos de entrenamiento)")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Sueldo (en $)")
plt.show()
```

6. Visualizacion los datos de TEST

```python
plt.scatter(x_test, y_test, color="red")
plt.plot(x_train, regression.predict(x_train), color="blue")
plt.title("Sueldo vs A√±os de experiencia (Conjunto de datos de TESTING)")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Sueldo (en $)")
plt.show()
```
