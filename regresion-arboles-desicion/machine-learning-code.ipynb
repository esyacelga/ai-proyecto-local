```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

dataset = pd.read_csv("./src/notebooks/data/Position_Salaries.csv")
dataset.head()
```

Analisis de la columna Profit, con respecto a la frecuencia de datos

```python
dataset.Level.hist(bins=40)
```

```python
x = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

Ajsutar la regresion con el data set

```python
from sklearn.tree import DecisionTreeClassifier
regression = DecisionTreeClassifier(random_state=0)
regression.fit(x,y)
```

```python
y_pred = regression.predict([[6.5]])
print(y_pred)
```

```python
import numpy as np
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x, y, color="red")
plt.plot(x, regression.predict(x), color="blue")
plt.title("Sueldo vs Años de experiencia (Conjunto de datos de TESTING)")
plt.xlabel("Años de experiencia")
plt.ylabel("Sueldo (en $)")
plt.show()
```
